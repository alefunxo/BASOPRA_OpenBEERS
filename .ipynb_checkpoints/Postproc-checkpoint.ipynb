```python
def func(row):
    if row.cluster<=2:
        return 'clow'
    elif row.cluster>=6:
        return 'ahigh'
    else:
        return 'bmedium'
    
```

```python
def plot_boxplot(df,col):
    '''
    Description
    -----------
    Boxplots of column 'col'. The Dataframe is grouped by pv penetration (pv) and battery penetration (batt) if not None.
    Parameters
    -----------
    df (pd.Series): Single row of the pd.Dataframe (for a single PV and Battery penetration)
    Returns
    -----------
    None
    '''
    size=16
    size2=12
    color=['blue', 'cyan','lime', 'red', 'green','orange']
    bp=df.round(0).boxplot(by=['house_type','classif'],  
                   column=col, 
                   grid=False)
#    tuples=[tuple(x) for x in df.loc[:,['house_type','cluster']].values]
    
    plt.draw()
    bp.set_ylabel(col,size=size)
    bp.set_yticklabels([item.get_text() for item in bp.get_yticklabels()],fontsize=size2)
    bp.set_xlabel('Type if house and cluster')
    bp.set_xticklabels([item.get_text() for item in bp.get_xticklabels()],fontsize=size2)

    bp.set_title('')
    plt.xticks( rotation='vertical')
    plt.suptitle('')
    if col=='bill' and not(pv and batt):
        bp.axhline(y=0, color='k')
```

Get the prices of DHW as if produced with the electricity (efficiency is 99% https://onlinelibrary.wiley.com/doi/epdf/10.1002/ese3.112)

```python
df_heat=pd.read_csv('Input/Heat_demand.csv',engine='python',sep=',|;',index_col=[0],
                    parse_dates=[0],infer_datetime_format=True )
df_heat.index=df_heat.index.tz_localize('UTC').tz_convert('Europe/Brussels')
filename_prices=Path('Input/Prices.csv')
df_prices=pd.read_csv(filename_prices,engine='python',sep=',|;',index_col=[0],
                            parse_dates=[0],infer_datetime_format=True )
if np.issubdtype(df_prices.index.dtype, np.datetime64):
    df_prices.index=df_prices.index.tz_localize('UTC').tz_convert('Europe/Brussels')
else:
    df_prices.index=pd.to_datetime(df_prices.index,utc=True)
    df_prices.index=df_prices.index.tz_convert('Europe/Brussels')
df_prices['DHW']=df_heat.DHW_kWh/.99
DHW_cost_DT=(df_prices.DHW*df_prices.Price_DT).sum()
DHW_cost_DT_mod=(df_prices.DHW*df_prices.Price_DT_mod).sum()
```

Moreover, the DHW may be provided using :

Electricity w/o tank with 99% efficiency and the same electricity prices used for the optimization.

Gas with a boiler of 90% efficiency and gas price of 0.0984 CHF/kWh

Oil with a boiler of 90% efficiency and gas price of 0.0828 CHF/kWh https://www.avenergy.ch/fr/prix-statistiques/prix/comparaison-des-prix-moyenne-mensuelle-et-annuelle


```python
DHW_cost_gas=(df_heat.DHW_kWh/.9*.0984).sum()
DHW_cost_oil=(df_heat.DHW_kWh/.9*.0828).sum()
```

# First let's calculate the results when no DPS

```python
filename=Path('Output/aggregated_results.csv')
dfa=pd.read_csv(filename,sep=';|,',engine='python')
filename=Path('Input/clusters.csv')
clusters=pd.read_csv(filename,sep=';|,',engine='python',index_col=[0])
dfa.cluster=dfa.apply(lambda x: clusters[x.hh==clusters.name].cluster.item() if clusters[x.hh==clusters.name].cluster.empty==False else -1,axis=1)
dfa['classif']=dfa.apply(lambda x: func(x),axis=1)
dfa['Total_bill_gas']=dfa.Bill
dfa['Total_bill_oil']=dfa.Bill
dfa.loc[(dfa.conf!=2)&(dfa.conf!=5),'Total_bill_gas']+=DHW_cost_gas #conf 2 and 5 include DHW with HP
dfa.loc[(dfa.conf!=2)&(dfa.conf!=5),'Total_bill_oil']+=DHW_cost_oil #conf 2 and 5 include DHW with HP
dfa_LS=dfa[(dfa.quartile==50)&(dfa.cluster!=-1)&(dfa.App_comb==2)&(dfa.Capacity==7)]
dfa_PS=dfa[(dfa.quartile==50)&(dfa.cluster!=-1)&(dfa.App_comb==3)&(dfa.Capacity==7)]
```
